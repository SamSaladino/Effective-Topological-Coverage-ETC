```python
def get_chebi(d):
    ref = d.get("reference")
    if isinstance(ref, str) and ref.lower().startswith("chebi:"):
        return f"CHEBI:{ref.split(':', 1)[1]}"
    for x in d.get("xrefs", []) or []:
        if isinstance(x, str) and x.lower().startswith("chebi:"):
            return f"CHEBI:{x.split(':', 1)[1]}"
    return None
```

#### Merge all JSON files in data/generated/metanetx (or pick first 3 with json_files[:3])

```python
json_files = sorted(metanetx_dir.glob("*.json"))
if not json_files:
    raise FileNotFoundError(f"No JSON files found in {metanetx_dir}")

merged = {}
for p in json_files:  # use [:3] to merge only three
    with open(p, "r") as f:
        d = json.load(f)
    if isinstance(d, dict):
        merged.update(d)
```

```python
def pubchem_smiles_from_inchikey(ik, timeout=10):
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/inchikey/{ik}/property/CanonicalSMILES/JSON"
    r = requests.get(url, timeout=timeout)
    if r.ok:
        props = r.json().get("PropertyTable", {}).get("Properties", [])
        if props:
            return props[0].get("CanonicalSMILES")
    return None
```

```python
def pubchem_smiles_from_chebi(chebi_id, timeout=10):
    # Accept "CHEBI:15377" or "15377"
    if not chebi_id:
        return None
    num = chebi_id.split(":")[-1]
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/sourceid/ChEBI/{num}/property/CanonicalSMILES/JSON"
    r = requests.get(url, timeout=timeout)
    if r.ok:
        props = r.json().get("PropertyTable", {}).get("Properties", [])
        if props:
            return props[0].get("CanonicalSMILES")
    return None
```

```python
def cactus_smiles(identifier, timeout=10):
    # identifier can be an InChIKey or "CHEBI:xxxx"
    url = f"https://cactus.nci.nih.gov/chemical/structure/{requests.utils.quote(identifier)}/smiles"
    r = requests.get(url, timeout=timeout)
    return r.text.strip() if r.ok and r.text else None
```

#### Fill missing SMILES using InChIKey -> PubChem -> Cactus

```python
mask_missing_smiles = df_all["smiles"].isna()
for idx in df_all[mask_missing_smiles].index:
    ik = df_all.at[idx, "inchikey"]
    smi = pubchem_smiles_from_inchikey(ik) or cactus_smiles(ik)
    if smi:
        df_all.at[idx, "smiles"] = smi
```

#### If CHEBI present but SMILES still missing, try CHEBI -> PubChem -> Cactus

```python
mask_missing_smiles = df_all["smiles"].isna() & df_all["chebi"].notna()
for idx in df_all[mask_missing_smiles].index:
    chebi = df_all.at[idx, "chebi"]
    smi = pubchem_smiles_from_chebi(chebi) or cactus_smiles(chebi)
    if smi:
        df_all.at[idx, "smiles"] = smi
```

### Method from metanetx

```python
def fill_chebi_from_metanetx_online(df: pd.DataFrame, url: str = "https://www.metanetx.org/cgi-bin/mnxget/mnxref/chem_xref.tsv", session: requests.Session | None = None) -> int:
    """
    If df['chebi'] is missing and df['metanetx'] has an MNXM id, fetch CHEBI from MetaNetX xrefs.
    Returns number of rows filled.
    """
    if "chebi" not in df.columns or "metanetx" not in df.columns:
        return 0

    sess = session or requests.Session()
    r = sess.get(url, timeout=30)
    r.raise_for_status()

    # Build MNXM -> CHEBI mapping
    mnx_to_chebi: dict[str, str] = {}
    for line in r.text.splitlines():  # robust to bytes issues
        if not line or line.startswith("#"):
            continue
        parts = line.split("\t")
        if len(parts) < 2:
            continue
        xref = parts[0].strip().lower()      # e.g., chebi:15377
        mnx_id = parts[1].strip().upper()    # e.g., MNXM1
        if xref.startswith("chebi:") and mnx_id.startswith("MNXM"):
            num = xref.split(":")[1]
            if num.isdigit():
                mnx_to_chebi[mnx_id] = f"CHEBI:{num}"

    # Apply mapping
    mask = df["chebi"].isna() & df["metanetx"].notna()
    filled = 0
    for idx in df[mask].index:
        mnx = str(df.at[idx, "metanetx"]).strip()
        # tolerate inputs like 'metanetx.chemical:MNXM123'
        if ":" in mnx and not mnx.upper().startswith("MNXM"):
            mnx = mnx.split(":")[-1]
        mnx = mnx.upper()
        chebi = mnx_to_chebi.get(mnx)
        if chebi:
            df.at[idx, "chebi"] = chebi
            filled += 1
    return filled
```

```python
before = int(df_human1["chebi"].isna().sum())
print(f"Missing CHEBI before: {before}")

filled = fill_chebi_from_metanetx_online(df_human1)
after = int(df_human1["chebi"].isna().sum())

print(f"Filled from MetaNetX by MNXM: {filled}")
print(f"Missing CHEBI after: {after}")
```

```python
import re
def fill_chebi_from_vmh_bigg_online(df: pd.DataFrame, session: requests.Session | None = None, timeout: int = 10) -> int:
    """
    If df['chebi'] is missing and df['vmhmetabolite'] has a value, try to fetch CHEBI using the BiGG universal metabolite API.
    Returns number of rows filled.
    """
    if "chebi" not in df.columns or "vmhmetabolite" not in df.columns:
        return 0

    sess = session or requests.Session()
    cache: dict[str, str | None] = {}
    filled = 0

    def normalize_candidates(vmh_id: str) -> list[str]:
        vmh_id = vmh_id.strip()
        # tolerate values like "vmhmetabolite:glc__D"
        if ":" in vmh_id and not vmh_id.startswith("MNXM"):
            vmh_id = vmh_id.split(":", 1)[-1]
        cands = [vmh_id]
        # try stripping common compartment suffixes like _c, _m, _e, etc.
        m = re.match(r"(.+)_([a-z]{1,2})$", vmh_id)
        if m:
            cands.append(m.group(1))
        # de-duplicate, keep order
        seen, ordered = set(), []
        for c in cands:
            if c and c not in seen:
                seen.add(c)
                ordered.append(c)
        return ordered

    mask_rows = df["chebi"].isna() & df["vmhmetabolite"].notna()
    for idx in df[mask_rows].index:
        raw = str(df.at[idx, "vmhmetabolite"])
        for candidate in normalize_candidates(raw):
            if candidate in cache:
                chebi = cache[candidate]
            else:
                chebi = None
                url = f"http://bigg.ucsd.edu/api/v2/universal/metabolites/{candidate}"
                try:
                    r = sess.get(url, timeout=timeout)
                    if r.ok:
                        j = r.json()
                        links = j.get("database_links") or {}
                        # find CHEBI entry case-insensitively
                        for k, v in links.items():
                            if k.lower() == "chebi" and isinstance(v, list) and v:
                                raw_id = v[0].get("id") or v[0].get("link") or ""
                                m = re.search(r"(?:CHEBI|ChEBI)\s*[:_ ]\s*(\d+)", str(raw_id))
                                if m:
                                    chebi = f"CHEBI:{m.group(1)}"
                                elif str(raw_id).isdigit():
                                    chebi = f"CHEBI:{raw_id}"
                                break
                except Exception:
                    chebi = None
                cache[candidate] = chebi

            if chebi:
                df.at[idx, "chebi"] = chebi
                filled += 1
                break  # stop trying candidates for this row
    return filled
```

```python
# Method 1: MetaNetX by MNXM
missing_start = int(df_human1["chebi"].isna().sum())
filled_mnx = fill_chebi_from_metanetx_online(df_human1)
missing_after_mnx = int(df_human1["chebi"].isna().sum())

# Method 2: VMH -> BiGG -> CHEBI
filled_vmh = fill_chebi_from_vmh_bigg_online(df_human1)
missing_after_both = int(df_human1["chebi"].isna().sum())

print(f"Missing CHEBI at start: {missing_start}")
print(f"Filled from MetaNetX: {filled_mnx}  | Remaining: {missing_after_mnx}")
print(f"Filled from VMH/BiGG: {filled_vmh}  | Remaining after both: {missing_after_both}")
```

```python
from libchebipy import ChebiEntity

# Function to get InChIKey from a ChEBI ID
def chebi_to_inchikey(chebi_id):
    try:
        if pd.notna(chebi_id):
            entity = ChebiEntity(chebi_id)
            return entity.get_inchi_key()
    except Exception as e:
        print(f"Error processing {chebi_id}: {e}")
    return None

# Create a list of InChIKeys from the ChEBI column
inchikeys_from_chebis = [
    chebi_to_inchikey(chebi_id) for chebi_id in df_human1["chebi"].dropna().unique()
]

# Remove None values and ensure uniqueness
inchikeys_from_chebis = list(filter(None, inchikeys_from_chebis))

# Compare with all_inchikeys
missing_inchikeys = set(all_inchikeys) - set(inchikeys_from_chebis)

print(f"Total InChIKeys from ChEBI: {len(inchikeys_from_chebis)}")
print(f"Total InChIKeys in all_inchikeys: {len(all_inchikeys)}")
print(f"Missing InChIKeys: {len(missing_inchikeys)}")

# Optionally, display missing InChIKeys
print("Missing InChIKeys:", missing_inchikeys)
```

```python
# Calculate the missing InChIKeys
missing_inchikeys = set(all_inchikeys) - set(inchikeys_from_chebis)

# Print the count of missing InChIKeys
print(f"Number of InChIKeys in all_inchikeys not in inchikeys_from_chebis: {len(missing_inchikeys)}")
```

```python
# Extract the part of the InChIKey before the first '-'
def extract_prefix(inchikey):
    return inchikey.split('-')[0] if inchikey and isinstance(inchikey, str) else None

# Apply the prefix extraction to both lists
inchikeys_from_chebis_prefix = {extract_prefix(ik) for ik in inchikeys_from_chebis}
all_inchikeys_prefix = {extract_prefix(ik) for ik in all_inchikeys}

# Compare the two sets
missing_prefixes = all_inchikeys_prefix - inchikeys_from_chebis_prefix

# Print the results
print(f"Total unique prefixes in all_inchikeys: {len(all_inchikeys_prefix)}")
print(f"Total unique prefixes in inchikeys_from_chebis: {len(inchikeys_from_chebis_prefix)}")
print(f"Missing prefixes: {len(missing_prefixes)}")
```
