```python
import networkx as nx
import matplotlib.pyplot as plt
from pathlib import Path
from etc.utils import GraphLabelPropagation 
```

```python
# Get the notebook's directory and go up to project root
notebook_dir = Path().resolve()
project_root = notebook_dir.parent
gml_path = project_root / "data" / "resources" / "generated" / "modified_graph.gml"
```

```python
G = nx.read_gml(gml_path)
```

```python
path = "/home/scostagonza/Documents/MappingEdata/mapping-data1/resources/nonEstelle.txt"
```

```python
# Load non-identified metabolites from txt file
non_identified_path = Path(path)
with open(non_identified_path, "r") as f:
    non_identified = {line.strip() for line in f if line.strip()}
```

```python
len(non_identified)
```

```python
# Set labels: 0 for non-identified, 1 for others
labels = {node: 0 if node in non_identified else 1 for node in G.nodes()}
glp = GraphLabelPropagation(G)
glp.labels = labels
```

```python
# Count how many nodes have value 1 and how many have value 0
from collections import Counter

label_counts = Counter(labels.values())
print(f"Nodes with value 1: {label_counts[1]}")
print(f"Nodes with value 0: {label_counts[0]}")
```

```python
# After applying label propagation
lpa_labels = glp.custom_lpa(iterations=100)

# Count how many nodes have value 1 and how many have value 0 in the propagated labels
from collections import Counter

lpa_label_counts = Counter(lpa_labels.values())
print(f"Nodes with value 1 after LPA: {lpa_label_counts[1]}")
print(f"Nodes with value 0 after LPA: {lpa_label_counts[0]}")
```
