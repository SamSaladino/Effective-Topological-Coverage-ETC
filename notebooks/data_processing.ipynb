```python
import pandas as pd
from pathlib import Path
import json

data_dir = Path.cwd().parent.parent / "Effective-Topological-Coverage-ETC"/ "data" 
devices_DC03 = data_dir / "resources" / "PilotStudy_DC03"
```

#### List of CSV files

```python
csv_files = [
    "blood_map_hilic_neg.csv",
    "blood_map_hilic_pos.csv",
    "blood_map_RPLC_neg.csv",
    "blood_map_RPLC_pos.csv"
]
```

#### Load each CSV into a DataFrame and display the first few rows

```python
for csv_file in csv_files:
    file_path = devices_DC03 / csv_file
    if file_path.exists():
        df = pd.read_csv(file_path)
        print(f"\nPreview of {csv_file}:")
        display(df.head())
    else:
        print(f"File not found: {file_path}")
```

#### Collect all True target_inchikeys

```python
inchikey_sets = []
for csv_file in csv_files:
    file_path = devices_DC03 / csv_file
    if file_path.exists():
        df = pd.read_csv(file_path)

        # Mask where Annotation is True
        filtered = df[df['Annotation'] == True]
        
        # Collect the target_inchikey values
        inchikey_sets.append(set(filtered['target_inchikey']))
    else:
        print(f"File not found: {file_path}")
```

#### Merge all sets into one set of unique inchikeys

```python
all_inchikeys = set.union(*inchikey_sets)
```

```python
print(f"Total unique target_inchikeys with Annotation=True: {len(all_inchikeys)}")
#print(all_inchikeys)
```

#### Export inchikeys as a txt file

```python
output_dir = data_dir  / "resources" / "generated"
output_file = output_dir / "unique_inchikeys.txt"


# Write each inchikey on a new line
with open(output_file, "w") as f:
    for inchikey in sorted(all_inchikeys):
        f.write(f"{inchikey}\n")

print(f"Exported {len(all_inchikeys)} inchikeys to {output_file}")
```

```python
metanetx_dir = output_dir / "metanetx"
```

```python
def get_chebi(d):
    ref = d.get("reference")
    if isinstance(ref, str) and ref.lower().startswith("chebi:"):
        return f"CHEBI:{ref.split(':', 1)[1]}"
    for x in d.get("xrefs", []) or []:
        if isinstance(x, str) and x.lower().startswith("chebi:"):
            return f"CHEBI:{x.split(':', 1)[1]}"
    return None
```

#### Merge all JSON files in data/generated/metanetx (or pick first 3 with json_files[:3])

```python
json_files = sorted(metanetx_dir.glob("*.json"))
if not json_files:
    raise FileNotFoundError(f"No JSON files found in {metanetx_dir}")

merged = {}
for p in json_files:  # use [:3] to merge only three
    with open(p, "r") as f:
        d = json.load(f)
    if isinstance(d, dict):
        merged.update(d)
```

Optional: persist the merged JSON

```python
with open(metanetx_dir / "merged_metanetx.json", "w") as f:
    json.dump(merged, f, indent=2)
```

#### Build the DataFrame: inchikey, CHEBI, SMILES

```python
rows = []
for inchikey, payload in merged.items():
    if not isinstance(payload, dict):
        continue
    rows.append({
        "inchikey": inchikey,
        "chebi": get_chebi(payload),
        "smiles": payload.get("SMILES")
    })

df_all = pd.DataFrame(rows)
```
